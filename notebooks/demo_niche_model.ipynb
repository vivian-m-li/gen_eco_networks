```python
import networkx as nx
import matplotlib.pyplot as plt

from gen_eco_networks.models import NicheModel

# instantiate the model
model = NicheModel(n_species=20, connectance=0.15, seed=42)

# generate the web using the niche model
web = model.generate()

# draw the web with trophic levels as y-coordinates
levels = nx.trophic_levels(web)
pos = nx.spring_layout(web, seed=42)
for node, level in levels.items():
    pos[node][1] = level  # set y-coordinate to trophic level
plt.figure(figsize=(8, 6))
nx.draw_networkx(
    web,
    pos,
    with_labels=True,
    node_size=200,
    node_color="lightblue",
    arrowsize=5,
)
plt.axis("off")
plt.tick_params(left=True, bottom=False, labelleft=True, labelbottom=False)
plt.show()
```

```python
from gen_eco_networks.models import NicheModel
from gen_eco_networks.utils import draw_network

SEED = 30

# instantiate the model
model = NicheModel(n_species=20, connectance=0.15, seed=SEED)

# generate the web using the niche model
web, _ = model.generate()

# visualize the generated web
draw_network(web, seed=SEED)
```

```python
# save the generated web to a gml file
model.save(web, "../example_networks/niche_model")

# save the generated web as node, edge, and attribute files
model.save(web, "../example_networks/niche_model", format="csv")

# read the web from files
read_web = model.read("../example_networks/niche_model", format="pkl")

# visualize the web
draw_network(read_web, seed=SEED)
```

```python
from gen_eco_networks.models import SBM
from gen_eco_networks.utils import draw_network

# instantiate the model
model = SBM(
    n_species=20,
    n_blocks=3,
    n_binary_attributes=1,
    n_numeric_attributes=2,
    reciprocal_proportion=0.1,
    seed=SEED,
)

# generate the web using the SBM
web, _ = model.generate()

# visualize the generated web
draw_network(web)

# save the generated web to example_networks/SBM/
model.save(web, "../example_networks/sbm")
model.save(web, "../example_networks/sbm", format="csv")
```

```python
# generate and visualize train/test splits of the web
train_subgraph, test_subgraph = model.train_test_split(web, test_size=0.2)
draw_network(train_subgraph)
draw_network(test_subgraph)
```

```python
from gen_eco_networks.models import RGG
from gen_eco_networks.utils import draw_network

# instantiate the model
model = RGG(
    n_species=20,
    radius=1.0,
    assortative=True,
    n_binary_attributes=2,
    n_numeric_attributes=4,
    seed=SEED,
)

# generate the web using RGG
web, _ = model.generate()

# visualize the generated web
draw_network(web)

# save the generated web to example_networks/rgg/
model.save(web, "../example_networks/rgg")
model.save(web, "../example_networks/rgg", format="csv")
```

```python
from gen_eco_networks.models import PNM
from gen_eco_networks.utils import draw_network

# instantiate the model
model = PNM(
    n_species=20,
    connectance=0.15,
    rho=0.5,
    n_binary_attributes=2,
    n_numeric_attributes=4,
    attribute_correlation_sigma=0.05,  # strong correlatino between generated attributes and niche parameters
    seed=SEED,
)

# generate the web using PNM
web, params = model.generate()

# visualize the generated web
draw_network(web)

# save the generated web to example_networks/rgg/
model.save(web, "../example_networks/pnm")
model.save(web, "../example_networks/pnm", format="csv")
```

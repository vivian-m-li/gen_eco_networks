```python
from gen_eco_networks.models import NicheModel
from gen_eco_networks.utils import draw_network

SEED = 30

# instantiate the model
model = NicheModel(n_species=20, connectance=0.15, seed=SEED)

# generate the web using the niche model
web, _ = model.generate()

# visualize the generated web
draw_network(web, seed=SEED)
```

```python
# save the generated web to a gml file
model.save(web, "../example_networks/niche_model")

# save the generated web as node, edge, and attribute files
model.save(web, "../example_networks/niche_model", gml=False)

# read the web from files
read_web = model.read("../example_networks/niche_model", gml=True)

# visualize the web
draw_network(read_web, seed=SEED)
```
